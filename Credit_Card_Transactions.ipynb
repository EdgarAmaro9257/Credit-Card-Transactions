 # **Cargar los datos** #

```python
# Cargar los datos
df = pd.read_csv('credit_card_transactions.csv')
df
```

```python
# Ver las primeras filas del DataFrame
print(df.head())
```

```python
# Obtener información básica del DataFrame
print(df.info())
```

```python
# Descripción estadística básica
print(df.describe())
```

 # **Limpieza de datos** #

```python
#Limpieza de datos
# Verificar valores nulos
print(df.isnull().sum())
```

```python
# Eliminar filas con valores nulos
df = df.dropna()
```

```python
# Eliminar filas duplicadas
df = df.drop_duplicates()
```

```python
# Verificación de limpieza
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
# Convertir las columnas de fechas a datetime si es necesario
# Supongamos que hay una columna llamada 'transaction_date'
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])

# Verificar la conversión
print(df['trans_date_trans_time'].head())
```

```python
# Distribución de variables numéricas
numerical_columns = df.select_dtypes(include=[np.number]).columns
df[numerical_columns].hist(bins=15, figsize=(15, 9), layout=(3, 4))
plt.show()
```

 # **Visualización de datos** #

```python
# Seleccionar solo columnas numéricas
numerical_df = df.select_dtypes(include=[np.number])

# Calcular la matriz de correlación
correlation_matrix = numerical_df.corr()
correlation_matrix
```

```python
#Visualización de datos
# Correlación entre variables numéricas

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Este gráfico ayuda a entender cómo están distribuidos los montos de las transacciones
# y si hay transacciones atípicamente altas que podrían ser fraudulentas.
plt.figure(figsize=(10, 6))
sns.histplot(df['amt'], bins=30, kde=True)
plt.title('Distribución del Monto de las Transacciones')
plt.xlabel('Monto ($)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Este gráfico muestra en qué horas del día ocurren más transacciones,
# lo que podría revelar patrones de fraude en horarios inusuales.
df['hour'] = df['trans_date_trans_time'].dt.hour
plt.figure(figsize=(10, 6))
sns.countplot(x='hour', data=df)
plt.title('Distribución de Transacciones por Hora del Día')
plt.xlabel('Hora del Día')
plt.ylabel('Número de Transacciones')
plt.show()
```

```python
# Este gráfico permite visualizar la distribución geográfica de las transacciones,
# destacando posibles concentraciones de fraude en ciertas áreas.

plt.figure(figsize=(10, 6))
sns.scatterplot(x='long', y='lat', hue='is_fraud', data=df)
plt.title('Distribución Geográfica de las Transacciones')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.show()
```

```python
# Este gráfico permite comparar la distribución de los montos de las transacciones entre las
# transacciones legítimas y fraudulentas, para ver si los montos son un indicador de fraude.

plt.figure(figsize=(10, 6))
sns.boxplot(x='is_fraud', y='amt', data=df)
plt.title('Boxplot del Monto de las Transacciones por Tipo de Fraude')
plt.xlabel('Fraude (0 = No, 1 = Sí)')
plt.ylabel('Monto ($)')
plt.show()
```

```python
# Este gráfico ayuda a identificar si ciertas categorías de comerciantes tienen
# una mayor incidencia de transacciones fraudulentas.
plt.figure(figsize=(10, 6))
sns.countplot(y='category', hue='is_fraud', data=df, order=df['category'].value_counts().index)
plt.title('Distribución de Transacciones por Categoría del Comerciante')
plt.xlabel('Número de Transacciones')
plt.ylabel('Categoría del Comerciante')
plt.show()
```
